# **Importaci칩n de datos**

```python
def get_txt_data(path_dict):
  out = []
  for k, v in path_dict.items():
    with open(v) as f:
      k = [line.rstrip() for line in f]
      out.append(k)
  return out
```

```python
def get_csv_data(path_dict):
  out = []
  for k, v in path_dict.items():
      k = pd.read_csv(v, index_col=[0])
      out.append(k)
  return out
```
